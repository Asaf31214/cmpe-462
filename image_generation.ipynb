```python
import os
import random
import torch
from diffusers import StableDiffusion3Pipeline
```

```python
model_id = "stabilityai/stable-diffusion-3.5-medium"
device = "cuda"
dtype = torch.bfloat16
```

```python
pipe = StableDiffusion3Pipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
    use_safetensors=True,
    safety_checker=None,
    requires_safety_checker=False
).to(device)
```

```python
items = ["banana", "carrot", "cucumber", "mandarina", "tomato"]
```

```python
quantities = ["a", "a", "a", "a", "a few"]
surfaces = ["table", "plate", "countertop", "cutting board", "flat surface"]
lightings = ["natural", "soft", "bright", "warm", "neutral"]
views = ["side", "top-down", "45-degree angle", "close-up"]
```

```python
def generate_prompts(item, count):
  prompts = []
  for _ in range(count):
    quantity = random.choice(quantities)
    surface = random.choice(surfaces)
    lighting = random.choice(lightings)
    view = random.choice(views)
    prompts.append(f"{quantity} {item}{"s" if quantity == "a few" else ""} on a {surface}, with {lighting} lighting, {view} view")
  return prompts
```

```python
generator = torch.Generator(device=device)
def generate_images(prompts):
  negative_prompt = "blurry, deformed, mutated, unrealistic colors, watermark, text, logo, reflections, mold, rotten, spoiled, bruised, discoloration, defects"
  negative_prompts = [negative_prompt] * len(prompts)
  return pipe(
    prompt=prompts,
    negative_prompt=negative_prompts,
    width=512,
    height=512,
    num_inference_steps=30,
    guidance_scale=4.5,
    generator=generator
).images
```

```python
image_per_item = 1000
batch_size = 25
```

```python
images = {item: [] for item in items}
for item in items:
  for start in range(0, image_per_item, batch_size):
    current_batch_size = min(batch_size, image_per_item - start)
    prompts = generate_prompts(item, current_batch_size)
    images[item].extend(generate_images(prompts))
```

```python
base_path = "drive/MyDrive/462-images"
for item_name, item_images in images.items():
  item_dir = os.path.join(base_path, item_name)
  os.makedirs(item_dir, exist_ok=True)
  for i, img in enumerate(item_images):
    img.save(os.path.join(item_dir, f"{i:03d}.png"))
```
