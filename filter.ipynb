```python
import os
import polars as pl
import numpy as np
from IPython.display import display
```

```python
df = pl.read_csv(os.path.join("data", "tabular", "feature_extraction.csv"))

y = df["class"]
X = df.drop("class")

X_np = X.to_numpy()
labels = y.to_list()
```

```python
keep_indices = []
classes = sorted(set(labels))
for cls in classes:
    idx = [i for i, c in enumerate(labels) if c == cls]
    X_cls = X_np[idx]
    centroid = X_cls.mean(axis=0)
    distances = np.linalg.norm(X_cls - centroid, axis=1)
    n_keep = len(idx) // 2
    keep_local = np.argsort(distances)[:n_keep]
    keep_indices.extend([idx[i] for i in keep_local])
```

```python
df_filtered = df[keep_indices]

output_path = os.path.join("data", "tabular", "feature_extraction_filtered.csv")
df_filtered.write_csv(output_path)
display(df_filtered)
```
